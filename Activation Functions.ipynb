#Sigmoid

```python
import math
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(x):
    return 1 / (1 + math.exp(-x)) # range 0 to 1
```

```python
sigmoid(100)
```

```python
#plotting the sigmoid function
x = np.linspace(-10, 10, 100)
y = [sigmoid(i) for i in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")
```

```python
sigmoid(-5)
```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x)) # range -1 to 1
```

```python
tanh(100)
```

```python
tanh(-5)
```

```python
#plotting the tanh function
x = np.linspace(-10, 10, 100)
y = [tanh(i) for i in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Tanh Function")
```

```python
def relu(x):
    return max(0,x) # range 0 to positive infinity
```

```python
relu(-5)
```

```python
#plotting the relu function
x = np.linspace(-10, 10, 100)
y = [relu(i) for i in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.title("Relu Function")
```

```python
def leaky_relu(x):
    return max(0.1*x,x) # range -0.1 to positive infinity
```

```python
leaky_relu(-5)
```

```python
#plotting the leaky relu function
x = np.linspace(-10, 10, 100)
y = [leaky_relu(i) for i in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("leaky_relu(x)")
plt.title("Leaky Relu Function")
```
