```python
import pandas as pd
```

```python
data = {"category":['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C']}
```

```python
df = pd.DataFrame(data)
df
```

```python
one_hot_encoded_df = pd.get_dummies(df,columns=['category']).astype(int)
one_hot_encoded_df
```

```python
one_hot_encoded_df = pd.get_dummies(df,columns=['category'],prefix="dummy",drop_first=True).astype(int)
one_hot_encoded_df
```
