```python
import pandas as pd
```

Data Load

```python
df = pd.read_csv('AQI and Lat Long of Countries.csv')
df.head()
```

Data Cleaning

```python
print(df["Country"].isnull().sum())
```

Dropped Missing Value

```python
df = df.dropna(subset=["Country"])
df.tail(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
df["AQI Value"].plot(
    kind="hist", ax=axes[0], color="skyblue", edgecolor="black", bins=20
)
df["lat"].plot(kind="hist", ax=axes[1], color="green", edgecolor="black", bins=20)
df["lng"].plot(
    kind="hist", ax=axes[2], color="orange", edgecolor="black", bins=20
)

axes[0].set_title("AQI Distribution")
axes[1].set_title("Latitude Distribution")
axes[2].set_title("Longitude Distribution")

plt.tight_layout()
plt.show()
```
